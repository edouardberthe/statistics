```python
x = np.linspace(-2, 2, 1001)
for (gen, name) in [(stats.weibull_min, "Weibull min"), (stats.weibull_max, "Weibull max")]:
    for param in [1, 3]:
        plt.plot(x, gen(param).pdf(x), label=f"{name}({param})")
plt.plot(x, stats.cauchy().pdf(x), label="Cauchy")
plt.legend()
plt.show()
```

```python
n = 100
x = stats.norm.ppf(np.arange(1, n+1) / (n+1))
for (gen,name) in [(stats.norm(), "Normal"), (stats.uniform(), "Uniform"), (stats.cauchy(), "Cauchy"), (stats.expon(), "Exponential")]:
    X = gen.rvs(n)
    X.sort()
    ax = plt.figure(figsize=(10, 6)).gca()
    ax.plot(x, X, ".", label=name)
plt.legend()
```
